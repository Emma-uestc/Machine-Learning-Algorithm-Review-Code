# Simulate Gradient Decent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,6,141)
```

```python
y = (x -2)**2 -1
```

```python
plt.plot(x,y)
plt.show()
```

```python
epsilon = 1e-8
alpha = 0.1
```

```python
def J(theta):
    return (theta - 2)**2 -1
def dJ(theta):
    return 2*(theta - 2)

theta = 0.0
while True:
    d_theta = dJ(theta)
    last_theta = theta
    theta = theta - alpha*d_theta
    
    if(abs(J(theta) - J(last_theta))<epsilon):
        break
        
print(theta)
print(J(theta))
```

```python
theta = 0.0
theta_lst = [theta]
while True:
    d_theta = dJ(theta)
    last_theta = theta
    theta = theta - alpha * d_theta
    theta_lst.append(theta)
    
    if(abs(J(theta) - J(last_theta))<epsilon):
        break
plt.plot(x,J(x))
plt.plot(np.array(theta_lst),J(np.array(theta_lst)),color = 'r',marker = '+')
plt.show()
```

```python
len(theta_lst)
```

```python
theta_lst = []
def gradient_decent(init_theta,alpha,epsilon= 1e-8):
    theta = init_theta
    theta_lst.append(init_theta)
    
    while True:
        d_theta = dJ(theta)
        last_theta = theta
        theta = theta - alpha * d_theta
        theta_lst.append(theta)
        
        if(abs(J(theta) - J(last_theta))<epsilon):
            break
            
def plot_theta_lst():
    plt.plot(x,J(x))
    plt.plot(np.array(theta_lst),J(np.array(theta_lst)),color = 'r',marker = '+')
    plt.show()
```

```python
alpha = 0.01
theta_lst = []
gradient_decent(0,alpha)
plot_theta_lst()
```

```python
def J(theta):
    try:
        return (theta - 2.)**2 -1.
    except:
        return float('inf')
```

```python
def gradient_decent(init_theta,alpha,n_iters = 1e3,epsilon = 1e-8):
    theta = init_theta
    i_iter = 0
    theta_lst.append(init_theta)
    
    while i_iter < n_iters:
        d_theta = dJ(theta)
        last_theta = theta
        theta = theta - alpha * d_theta
        theta_lst.append(theta)
        
        if(abs(J(theta) - J(last_theta))<epsilon):
            break
            i_iter += 1
            
    return
```

```python
alpha = 1.1
theta_lst = []
gradient_decent(0,alpha)
```
